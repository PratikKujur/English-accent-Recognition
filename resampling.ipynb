```python
import os
import librosa
import soundfile as sf
```

```python
data_directory = 'GLOBE\wavfiles'
target_sample_rate = 16000
```

```python
def resample_audio(file_path, target_sr):
    try:
        # Load the audio file
        audio, sr = librosa.load(file_path, sr=None)  # Load with original sample rate
        # Resample the audio to the target sample rate
        audio_resampled = librosa.resample(audio, orig_sr=sr, target_sr=target_sr)
        # Save the resampled audio back to the file
        sf.write(file_path, audio_resampled, target_sr)
        print(f"Resampled and saved: {file_path}")
    except Exception as e:
        print(f"Error processing {file_path}: {e}")
```

```python
for accent_folder in os.listdir(data_directory):
    accent_path = os.path.join(data_directory, accent_folder)
    if os.path.isdir(accent_path):
        audio_files = [f for f in os.listdir(accent_path) if f.endswith('.wav')]

        # Iterate over each audio file in the accent folder
        for audio_file in audio_files:
            file_path = os.path.join(accent_path, audio_file)
            
            # Load the audio file
            try:
                audio, sr = librosa.load(file_path, sr=None)  # sr=None to preserve original sample rate
                if sr != target_sample_rate:
                    print(f"Resampling {file_path} from {sr}Hz to {target_sample_rate}Hz")
                    
                    # Resample the audio to the target sample rate
                    audio_resampled = librosa.resample(audio, orig_sr=sr, target_sr=target_sample_rate)

                    # Save the resampled audio, replacing the original file
                    sf.write(file_path, audio_resampled, target_sample_rate)
                else:
                    print(f"Skipping {file_path} as it already has the target sample rate {target_sample_rate}Hz")
            except Exception as e:
                print(f"Error processing {file_path}: {e}")
```
